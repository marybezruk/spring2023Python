```python
import numpy as np
import numpy.linalg
import numpy.random
import matplotlib.pyplot as plt
```

# Номер 1

```python
def f(x):
    return np.sin(x/5) * np.exp(x/10) + 5 * np.exp(-1/2 * x)


def approximation(n):
    a = np.linspace(1, 15, n)
    A = np.array([[x ** i for i in range(n)] for x in a])
    B = f(a)
    W = np.linalg.solve(A, B)
    y = np.linspace(1,15,250)
    answer = np.array(sum([W[i] * y**i for i in range(n)]))
    return answer


xx = np.linspace(1, 15, 250)
n = 3
xx2 = np.linspace(1, 15, n)
plt.grid()
plt.xlabel('Оx')
plt.ylabel('Оy')
plt.plot(xx, f(xx), label = 'Исходная функция')
plt.plot(xx2, f(xx2), marker = '.', linestyle = '', label = f'Точки для аппроксимации при n = {n}')
plt.legend()
plt.show()

plt.grid()
plt.xlabel('Оx')
plt.ylabel('Оy')
plt.plot(xx, approximation(n), label = f'Аппроксимация при n = {n}')
plt.plot(xx2, f(xx2), marker = '.', linestyle = '', label = f'Точки для аппроксимации при n = {n}')
plt.legend()
plt.show()
```

# Номер 2

```python
def noize_approximation(n):
    a = np.linspace(1, 15, n)
    A = np.array([[x ** i for i in range(n)] for x in a])
    B = f(a)
    for i in range(len(B)):
        B[i] += np.random.normal(0, 0.01)
    W = np.linalg.solve(A, B)
    y = np.linspace(1,15,250)
    answer = np.array(sum([W[i] * y**i for i in range(n)]))
    return answer, B


xx = np.linspace(1, 15, 250)
n = 5
xx2 = np.linspace(1, 15, n)
yy = noize_approximation(n)
plt.grid()
plt.xlabel('Оx')
plt.ylabel('Оy')
plt.plot(xx, f(xx), label = 'Исходная функция')
plt.plot(xx2, yy[1], marker = '.', linestyle = '', label = f'Точки для аппроксимации при n = {n}')
plt.legend()
plt.show()

plt.grid()
plt.xlabel('Оx')
plt.ylabel('Оy')
plt.plot(xx, yy[0], label = f'Аппроксимация при n = {n}')
plt.plot(xx2, yy[1], marker = '.', linestyle = '', label = f'Точки для аппроксимации при n = {n}')
plt.legend()
plt.show()
```

# Номер 3

```python
def f(x):
    return np.sin(x/5) * np.exp(x/10) + 5 * np.exp(-1/2 * x)


def least_squares(n):
    N = 100
    a = np.linspace(1, 15, N)
    A = np.array([[x ** i for i in range(n)] for x in a])
    B = f(a)
    for i in range(len(B)):
        B[i] += np.random.normal(0, 0.1)
    W, unnes1, unnes2, unnes3 = np.linalg.lstsq(A, B, rcond=None)
    answer = np.array([sum([W[i] * x**i for i in range(n)]) for x in a])
    return answer, B


xx = np.linspace(1, 15, 250)
N = 100
xx2 = np.linspace(1, 15, N)
n = 4 # так как полином степени n - 1 => если нужен кубический полином, то n = 4
yy = least_squares(n)
plt.grid()
plt.xlabel('Оx')
plt.ylabel('Оy')
plt.plot(xx, f(xx), label = 'Исходная функция')
plt.plot(xx2, yy[1], marker = '.', linestyle = '', label = f'Точки для аппроксимации при n = {n}')
plt.legend()
plt.show()

plt.grid()
plt.xlabel('Оx')
plt.ylabel('Оy')
plt.plot(xx2, yy[0], label = f'Аппроксимация при n = {n}')
plt.plot(xx2, yy[1], marker = '.', linestyle = '', label = f'Точки для аппроксимации при n = {n}')
plt.legend()
plt.show()
```

# Выводы

Полиномиальная регрессия хорошо аппроксимирует функции при больших n и отсутствии сильного шума, метод наименьших квадратов же аппроксимирует довольно точно и при маленьких n и сильном шуме
