## Exercise 1  
1. Создайте график $x^2⋅sin(1/x^2)+x$ на интервале $[−1,1]$, используя 250 точек. Не забудьте пометить оси!
2. Создайте semilogy график (с логарифмической шкалой для оси y и линейной шкалой для оси x) относительной разницы  функций $1/(1+x^2)$ и $1/x^2$ на интервале $[5,25]$. (Относительная разница чисел $a$ и $b$ определяется как $|1−a/b|$)
3. В Scipy также есть много менее распространенных специальных функций. Взгляните на обширный список <a href=https://docs.scipy.org/doc/scipy/reference/special.html> здесь</a>! Выберите несколько интересных спец функций, которые вы знаете (или, может быть, не знаете), подставьте несколько чисел и начертите их.


```python
xx = np.linspace(-1, 1, 250)
yy = xx**2 * np.sin(1/xx**2) + xx
plt.title('График 1')
plt.xlabel(u'Ox', family='verdana')
plt.ylabel(u'Oy', family='verdana')
plt.plot(xx, yy)
```

```python
xx = np.linspace(5, 25, 75)
aa = 1 / (1 + xx ** 2)
bb = 1 / xx ** 2
yy = abs(1 - aa / bb)
plt.title('График 2')
plt.xlabel(u'Ox', family='verdana')
plt.ylabel(u'Ln(Oy)', family='verdana')
plt.semilogy(xx, yy)
```

```python
xx = np.linspace(-5, 5, 50)
yy = scipy.special.gamma(xx) / 25 # на максимум делить не хочет, поэтому нормировка условная
yy2 = scipy.special.psi(xx) / 50
yy3 = scipy.special.erf(xx)
plt.title('График 3')
plt.xlabel(u'Ox', family='verdana')
plt.ylabel(u'Oy', family='verdana')
plt.plot(xx, yy, label = 'нормированная гамма функция')
plt.plot(xx, yy2, label = 'нормированная пси функция', color = 'red', linestyle = '--')
plt.plot(xx, yy3, label = 'функция ошибки', color = 'black', linestyle = '', marker = '+')
plt.legend()
```

## Exercise 2  
1. Построить график функции $sin(x)/x$ на отрезке $[-6\pi;6\pi]$ и добавить тики по оси $x$ с шагом $\pi/2$ от $-6\pi$ до $6\pi$, причём обозначения тиков должны содержать греческую букву $\pi$. Добавить сетку на график.

```python
xx = np.linspace(-6*np.pi, 6*np.pi, 250)
yy = np.sin(xx) / xx
plt.figure(figsize=(12, 6))
plt.title('График 3')
plt.grid()
plt.xlabel(u'Ox', family='verdana')
plt.ylabel(u'Oy', family='verdana')
plt.plot(xx, yy, label = 'sin(x)/x')
xt = np.arange(-6*np.pi, 6*np.pi + np.pi/2, np.pi/2)
yt = []
for i in range(-12, 13):
    if i % 2 == 0:
        if i == 0:
            yt.append('0')
        elif i == -2:
            yt.append(r'$-\pi$')
        elif i == 2:
            yt.append(r'$\pi$')
        else:
            yt.append(r'$%s\pi$'%(i // 2))
    else:
        if i != 1 and i != -1:
            yt.append(r'$\dfrac{%s\pi}{2}$'%(i))
        elif i == -1:
            yt.append(r'$-\dfrac{\pi}{2}$')
        else:
            yt.append(r'$\dfrac{\pi}{2}$')
plt.xticks(xt, yt)
plt.legend()
```

## Exercise 3  
1. Создайте цветовой график $sin(x)sin(y)$ на интервале $[−π,π]×[−π,π]$.
2. Создайте цветовой график $e^{-x^2-y^2}$ на интервале $[−3,3]×[−3,3]$.

```python
x = np.linspace(-np.pi,np.pi,150)
y = np.linspace(-np.pi,np.pi,150)
X,Y = np.meshgrid(x,y)

Z = np.sin(X) * np.sin(Y)

fig, ax = plt.subplots(figsize=(6,5))

norm = mpl.colors.Normalize(-abs(Z).max(), abs(Z).max())

p = ax.pcolor(X, Y, Z, norm = norm, cmap='jet')
ax.axis('tight')
ax.xaxis.set_major_locator(mpl.ticker.MaxNLocator(4))
ax.yaxis.set_major_locator(mpl.ticker.MaxNLocator(4))

cb = fig.colorbar(p, ax=ax)
cb.set_ticks([-1, -.5, 0, .5, 1])
```

```python
x = np.linspace(-3,3,150)
y = np.linspace(-3,3,150)
X,Y = np.meshgrid(x,y)

Z = np.exp(-X**2 - Y**2)

fig, ax = plt.subplots(figsize=(6,5))

norm = mpl.colors.Normalize(-abs(Z).max(), abs(Z).max())

p = ax.pcolor(X, Y, Z, norm = norm, cmap='jet')
ax.axis('tight')
ax.xaxis.set_major_locator(mpl.ticker.MaxNLocator(4))
ax.yaxis.set_major_locator(mpl.ticker.MaxNLocator(4))

cb = fig.colorbar(p, ax=ax)
cb.set_ticks([-1, -.5, 0, .5, 1])
```
